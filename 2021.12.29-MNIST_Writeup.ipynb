MNIST Update
===

**Prerequisite**

You must have a copy of the MNIST training set in CSV format. You can find a copy to download [here](https://www.kaggle.com/oddrationale/mnist-in-csv)

Reading and converting CSV file to an array
---

```python
import pandas as pd
import numpy as np

res = pd.read_csv('mnist_train.csv') 
res_arr = np.array(res)

res

```

```python
# This array contains arrays of arrays that are equivalent to each column in the CSV file.

arr_split = np.split(res_arr, 785,1)
```

Processing Array for int 5
---

```python
# Isolate the array containing the number labels 
int_labels = arr_split[0] 
int_labels

# Implement logic to isolate the position of the number 5 in the array
int_5 = int_labels == 5

# Function to implement this logic in the other arrays
filter_vals = lambda x: x[int_5]

# List containing new arrays based on the position of 5 in the remaining columns

filter_for_5 = [] 
i = 1
while i < 785:
    filter_for_5.append(filter_vals(arr_split[i]))
    i += 1
```

Reorientate Array
---

```python
filter_for_5_arr = np.array(filter_for_5) # converting list to an array
```

```python
# The arrays previously corresponded to the values in column positon in the CSV file.
# Transposing the array adjusts the arrays to contain values associated with the rows
# in the CSV file.


filter_for_5_arr2 = np.transpose(filter_for_5_arr) 
```

Create Sample Image
--

```python
sample_img = filter_for_5_arr2[15]
```

```python
test_sample_enhance = np.split(sample_img, 28) 
# splitting array for displaying number correctly

df_img_test = pd.DataFrame(test_sample_enhance[4:50])

df_img_test
df_img_test.style.set_properties(**{'font-size': '6pt'}).background_gradient('Greys')
```

Solution - Generating mean
--

```python
# With the columns transposed, each row of the array corresponds to a complete 28x28 image.
# This function takes the mean of each value from each array i.e. the value from 1x1 in 
# the first array is combined with 1x1 in the second and so on until the mean can be 
# calculated

filter_for_5_mean_arr = np.mean(filter_for_5_arr2, axis = 0)
```

```python
new_test_sample_enhance = np.split(filter_for_5_mean_arr, 28)

new_df_img_test = pd.DataFrame(new_test_sample_enhance[4:50])

new_df_img_test
new_df_img_test.style.set_properties(**{'font-size': '6pt'}).background_gradient('Greys')
```
